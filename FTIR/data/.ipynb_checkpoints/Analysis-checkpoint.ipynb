```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# from matplotlib.ticker import MultipleLocator
from scipy.stats import linregress
```

```python
def readData(path): # Функция для считывания данных
    data = pd.read_table(path, sep = '\s+' )
    x = data['x']
    y = data['y']
    return(x,y)
```

```python
def plotData(x,y):
    plt.figure(figsize= (15,5)) 
    plt.plot(x, y,color = 'black')
    plt.xlabel("\nu$, " + "см^{-1}$")
    plt.ylabel("Интенсивность")
    plt.grid()
```

```python
path = '/home/dmitry/LatexProjects/physicalMethods/FTIR/data/task1_air_0_5.ascii'

x,y = readData(path)

plt.figure(figsize= (15,5)) 
plt.plot(x[:len(x)//2], y[:len(y)//2],color = 'black')
plt.xlabel("\nu$, " + "см^{-1}$")
plt.ylabel("Интенсивность")
plt.grid()

```
